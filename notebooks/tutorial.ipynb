### 3.1

````python
import math

# 3.1 Write a Python script to read quadratic equation coefficients and calculate solutions

def solve_quadratic(a, b, c):
    """
    Solves a quadratic equation ax^2 + bx + c = 0
    
    Args:
        a, b, c: coefficients of the quadratic equation
    
    Returns:
        tuple: solutions (x1, x2) or None if no real solutions
    """
    # Check if a = 0 (not a quadratic equation)
    if a == 0:
        print("Error: 'a' cannot be zero. This is not a quadratic equation.")
        return None
    
    # Calculate discriminant
    discriminant = b**2 - 4*a*c
    
    # Check if discriminant is negative (no real solutions)
    if discriminant < 0:
        print("Error: Discriminant is negative. No real solutions exist.")
        return None
    
    # Calculate solutions
    sqrt_discriminant = math.sqrt(discriminant)
    x1 = (-b + sqrt_discriminant) / (2*a)
    x2 = (-b - sqrt_discriminant) / (2*a)
    
    return (x1, x2)

# Get user input
print("=== Quadratic Equation Solver ===")
print("Solve: ax² + bx + c = 0\n")

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

print(f"\nEquation: {a}x² + {b}x + {c} = 0")

# Solve the equation
result = solve_quadratic(a, b, c)

if result is not None:
    x1, x2 = result
    print(f"\nSolutions:")
    print(f"x1 = {x1}")
    print(f"x2 = {x2}")
````

### 3.2 - When discriminant equals zero

**What happens when the discriminant b² - 4ac = 0?**

When the discriminant equals zero, the quadratic equation has exactly **one real solution** (a repeated root).
The formula becomes: $x = \frac{-b}{2a}$

**Example: a=1, b=-2, c=1 (equation: x² - 2x + 1 = 0)**
- Discriminant = (-2)² - 4(1)(1) = 4 - 4 = 0
- Solution: x = 2/2 = 1 (double root)
- This equation factors as (x - 1)² = 0

````python
# Example 3.2: Discriminant equals zero
# Equation: x² - 2x + 1 = 0

a_ex = 1
b_ex = -2
c_ex = 1

discriminant = b_ex**2 - 4*a_ex*c_ex
print(f"Example: {a_ex}x² + {b_ex}x + {c_ex} = 0")
print(f"Discriminant = {b_ex}² - 4({a_ex})({c_ex}) = {discriminant}")

if discriminant == 0:
    x = -b_ex / (2*a_ex)
    print(f"Since discriminant = 0, there is one unique solution (double root):")
    print(f"x = {x}")
````

### 3.3 - Handling negative discriminant

**What happens with a negative discriminant?**

If the discriminant is negative (b² - 4ac < 0), the solutions are **complex numbers** (not real numbers).

**Example: a=10, b=5, c=6**
- Discriminant = 5² - 4(10)(6) = 25 - 240 = -215
- Since the discriminant is negative, there are no real solutions
- Solution: We must check for this condition and report an error or handle complex numbers

````python
# Example 3.3: Negative discriminant
# Equation: 10x² + 5x + 6 = 0

a_ex = 10
b_ex = 5
c_ex = 6

discriminant = b_ex**2 - 4*a_ex*c_ex
print(f"Example: {a_ex}x² + {b_ex}x + {c_ex} = 0")
print(f"Discriminant = {b_ex}² - 4({a_ex})({c_ex}) = {discriminant}")

if discriminant < 0:
    print("Since the discriminant is negative, there are no real solutions.")
    print("The solutions would be complex numbers.")
    
    # Optionally, we can calculate complex solutions
    real_part = -b_ex / (2*a_ex)
    imag_part = math.sqrt(-discriminant) / (2*a_ex)
    print(f"\nComplex solutions:")
    print(f"x1 = {real_part} + {imag_part}i")
    print(f"x2 = {real_part} - {imag_part}i")
````

### 3.4 - Handling a = 0

**What happens when a = 0?**

If a = 0, the equation becomes: 0·x² + bx + c = 0, which simplifies to bx + c = 0

This is a **linear equation**, not a quadratic equation. Division by 2a in the quadratic formula would cause a **division by zero error**.

**Solution:** Check if a ≠ 0 before using the quadratic formula. If a = 0 and b ≠ 0, solve it as a linear equation: x = -c/b

### 3.5 - Combined error handling

**Combining both problems into a single IF statement:**

The two possible problems with the quadratic equation - negative discriminant and first coefficient equals to zero - can be handled by using a single IF statement:

```python
if a == 0 or d < 0:
    print('Error')
else:
    # calculate solutions
```

**Possible issue with this approach:**

This approach treats both error conditions the same way, but they are actually different problems:
- When **a = 0**, we have a linear equation (not quadratic), which may or may not have a solution depending on b
- When **discriminant < 0**, we have a quadratic equation but with complex solutions instead of real ones

A better approach would be to handle them separately to provide more informative error messages to the user.

````python
# Example 3.5: Combined error handling approach
# This demonstrates the combined IF statement as suggested in the exercise

def solve_quadratic_combined(a, b, c):
    """
    Solves quadratic equation with combined error handling.
    Shows both problems (a=0 and negative discriminant) together.
    """
    d = b**2 - 4*a*c  # Calculate discriminant
    
    if a == 0 or d < 0:
        print('Error: Cannot solve (either a=0 or discriminant is negative)')
        return None
    else:
        # Calculate solutions
        sqrt_d = math.sqrt(d)
        x1 = (-b + sqrt_d) / (2*a)
        x2 = (-b - sqrt_d) / (2*a)
        return (x1, x2)

# Test cases demonstrating both error conditions
print("Test 1: Normal case (a=1, b=-3, c=2)")
result = solve_quadratic_combined(1, -3, 2)
if result:
    print(f"Solutions: x1 = {result[0]}, x2 = {result[1]}\n")

print("Test 2: a = 0 (linear equation)")
result = solve_quadratic_combined(0, 2, 3)
print()

print("Test 3: Negative discriminant")
result = solve_quadratic_combined(10, 5, 6)
print()

print("\n=== Issue with combined approach ===")
print("By combining both error conditions in a single IF statement,")
print("we lose information about WHAT the error is:")
print("- Is it a linear equation (a=0)?")
print("- Or does it have complex solutions (negative discriminant)?")
print("\nA better approach would handle each error separately for clarity.")
````
